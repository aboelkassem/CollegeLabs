```python
import matplotlib.pyplot as plt
import numpy as np
```

**General sine wave equation**

image.png


```python
# equation parameters
freq = int(input("Enter the frequency: "))
phase = int(input("Enter the phase: "))
A = int(input("Enter the amplitude: "))

# sampling rate
sr = 100.0
# sampling interval
ts = 1.0/sr
t = np.arange(0,1,ts)

y = A * np.sin(2*np.pi*freq*t + phase)
```

**Ploting the sine wave**

```python
plt.figure(figsize = (8, 8))
plt.plot(t, y, 'b')
plt.ylabel('Amplitude')
plt.xlabel('Time (s)')
plt.grid(True, which='both')
plt.axhline(y=0, color='k')
plt.show()
```
